```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.columns
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3).fit(iris_df[['sepal length (cm)', 'sepal width (cm)']])
clusters=kmeans.predict(iris_df[['sepal length (cm)', 'sepal width (cm)']])
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('K-Means Clustering with k=3')
plt.legend(loc='upper right')
for i in range(0,150):
    if clusters[i]==0:
        plt.scatter(iris_df.at[i,'sepal length (cm)'],iris_df.at[i,'sepal width (cm)'],c='red', label='Cluster 0')
    if clusters[i]==1:
        plt.scatter(iris_df.at[i,'sepal length (cm)'],iris_df.at[i,'sepal width (cm)'],c='blue', label='Cluster 1')
    if clusters[i]==2:
        plt.scatter(iris_df.at[i,'sepal length (cm)'],iris_df.at[i,'sepal width (cm)'],c='green', label='Cluster 2')
plt.show()
```
